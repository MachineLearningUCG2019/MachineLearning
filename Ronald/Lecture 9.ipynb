```python
f = open('/Users/Ronald van Hal/My Documents/Python Scripts/Machine Learning/titanic.csv')

Survived = []
Pclass = []
Age = []

for i in f:
    try:
        t = i.split(',')
        if t[6] == '':
            continue
        else:
            Survived.append(int(t[1]))
            Pclass.append(int(t[2]))
            Age.append(float(t[6]))
    except:
        pass



true_positive = 0
false_positive = 0
false_negative = 0
true_negative = 0

T = [0,0,0,0]

#for i in range(len(Survived)):
#    Survived[i] = int(Survived[i])
#    Pclass[i] = int(Pclass[i])


for i in range(len(Survived)):
    
    if Age[i] <= 30.0:
        if Pclass[i] == 3:
            assumption = "Dead"
        elif Pclass[i] == 2:
            assumption = "Survived"
        elif Pclass[i] == 1:
            assumption = "Survived"
    elif Age[i] > 30.0:
        if Pclass[i] == 3:
            assumption = "Dead"
        elif Pclass[i] == 2:
            assumption = "Dead"
        elif Pclass[i] == 1:
            assumption = "Survived"

    if assumption == "Survived": 
        if Survived[i] == 1:
            T[0] += 1
            true_positive += 1
        elif Survived[i] == 0:
            T[1] += 1
            false_positive += 1
    elif assumption == "Dead":
        if Survived[i] == 1:
            T[2] += 1
            false_negative += 1
        if Survived[i] == 0:
            T[3] += 1
            true_negative += 1

Accuracy = (true_positive + true_negative) / sum(T)
print(T)
print(Accuracy)
```
