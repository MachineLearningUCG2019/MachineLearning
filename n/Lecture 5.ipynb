```python
def dF(y, x, a, b):
    T=y-(a*x+b)
    first=-2*x*T
    second=2*T
    return np.sum(first), np.sum(second)

dF(y,x,1,-2)
```

```python
def cost(y, x, a, b):
    cost=0.5*np.sum((y-(a*x+b))**2)/len(y)
    return cost

def derivative_values(y, x, a, b):
    T=(y-(a*x+b))
    first_a=0.5*(np.sum((-2)*x*T))/len(y)
    second_b=0.5*(np.sum(T*(-2)))/len(y)
    return first_a, second_b
    
def GradientDescent(y,x,a,b,n):
    hist=[]
    for i in range(n):
        Der_Value=derivative_values(y,x,a,b)
        a = a - 0.01*Der_Value[0]
        b = b - 0.01*Der_Value[1]
        hist.append((a,b))
    return a,b,hist

```

```python

a,b,hist=GradientDescent(y,x,5,1, 1000)
print(a,b,hist)
ym=a*x+b
plt.scatter(x,y)
plt.plot(x,ym)
plt.show()
fig, ax = plt.subplots()
#ax.scatter(x,y)
for i in range(len(hist)):
    ax.cla()
    ym=hist[i][0]*x+hist[i][1]
    ax.plot(x,ym)
    ax.scatter(x,y)
    ax.set_title(str(hist[i]))
    # Note that using time.sleep does *not* work here!
    plt.pause(0.01)
```
